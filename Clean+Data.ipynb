```python
import pandas as pd
data = pd.read_csv('US_Accidents_June20.csv')
data.head()
```

```python
len(data)
```

```python
data.isnull().sum(axis = 0)
```

```python
data['TMC'].fillna(value=201, inplace=True) 
```

```python
data['Precipitation(in)'] = data['Precipitation(in)'].fillna(data['Precipitation(in)'].median())
data['Visibility(mi)'] = data['Visibility(mi)'].fillna(data['Visibility(mi)'].median())
data['Temperature(F)'] = data['Temperature(F)'].fillna(data['Temperature(F)'].median())
data['Humidity(%)'] = data['Humidity(%)'].fillna(data['Humidity(%)'].median())
```

data.drop(['End_Lat', 'End_Lng', 'Number',"Timezone", "Airport_Code", "Weather_Timestamp", "Zipcode",'Wind_Chill(F)' ], inplace=True)
data.isnull().sum(axis = 0)

```python
data.loc[data['Wind_Direction'] == 'Calm', 'Wind_Speed(mph)'] = 0
```

```python
data.isnull().sum(axis = 0)
```

```python
data.drop(['Pressure(in)'], axis = 1)
```

```python
print('Missing values: ', data.columns[(100*data.isnull().sum()/data.shape[0])>0].tolist())
```

```python
data.isnull().sum(axis = 0)
```

```python
data.dropna(inplace=True)
```
